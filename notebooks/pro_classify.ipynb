```python
import subprocess

# Absolute or relative path to your Prolog file
prolog_main = "../src_prolog/main.pl"

# Construct the query: run Prolog with a goal to classify a parcel
query = "classify_parcel(p1)."

# Run SWI-Prolog
result = subprocess.run(
    ["swipl", "-q", "-s", prolog_main, "-g", query, "-t", "halt"],
    capture_output=True,
    text=True
)

# Output
print("STDOUT:\n", result.stdout)
print("STDERR:\n", result.stderr)
```
